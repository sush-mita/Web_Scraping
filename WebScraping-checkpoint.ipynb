# Extracting Ratings of the Shows and Movies

```python
div = docs.find_all('div' , {'class' :'lister-item-content' })
ran = len(div)
Reviews = []
for i in range(ran):
    result = div[i].find_all('strong')
    if (len(result) == 0):
        Reviews.append(' ')
    else:
        rat= div[i].find_all('strong')[0].text
        Reviews.append(rat)
    
Reviews
```
